```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# bar chart - compare categories of data by representing each category with a bar

categories = ['Grains','Fruits','Vegetables','Protein','Dairy','Sweet']
values = np.array([4,3,2,5,3,1])

# plt.bar(categories,values, color='skyblue')
plt.barh(categories,values, color='skyblue')

plt.title('Daily Consumption')
plt.xlabel('Food')
plt.ylabel('Quantity')
plt.show()
```

```python
# Pie chart - Circular chart divided into slices to show percentages of the total.
#             Good for visualizing distribution among categories. 

categories = ['Freshmen','Sophomores','Juniors','Seniors']
values = np.array([300,250,275,225])
color = ['red','yellow','blue','green']

plt.pie(values, labels=categories, autopct='%1.2f%%',colors = color, explode=[0,0,0,0.2], shadow=True, startangle=180)
plt.title('College')
plt.show()
```

```python
# Scatter Graph -  Shows the relationship between two variables
#                  Hepls to identify correlation (+, - , None)
#                  Ex: study hours vs  test scores

x = np.array([0,1,1,2,3,4,5,6,7,7,8])
y = np.array([55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87])

x2 = np.array([0,1,1,2,3,4,5,6,7,8,8])
y2 = np.array([50, 58, 65, 70, 72,78,83,88,92,95,97])

#plt.scatter((x,x2),(y,y2), color='skyblue', alpha=0.5, s=100)
plt.scatter(x,y, color='red', alpha=0.5, s=100, label='Class A')
plt.scatter(x2,y2, color='skyblue', alpha=0.5, s=100, label='Class B')
plt.title('Test Score')
plt.xlabel('Hours Study')
plt.ylabel('Grade')
plt.legend()
plt.show()
```

```python
#  Histogram - Visaul representation of the distribution of quantitative data
#              They group values into bins (intervals)
#              and counts how many fall in each range. 

scores = np.random.normal(loc=80, scale=10, size=100)
scores = np.clip(scores, 0, 100)

plt.hist(scores,bins=10, color='red', alpha=0.5,edgecolor='black')
plt.title('Exam Scores Distribution')
plt.xlabel('Score')
plt.ylabel('No. of Students')
plt.show()
```

```python
# Subplots - multiple plots in a single figure for comparison

# Figure = the entire canvas
# Ax = a single plot(subplot) within the figure

x = np.array([1,2,3,4,5])


figure, axes = plt.subplots(3,3)  # 3 rows, 3 columns
# figure, axes = plt.subplots(1,3)  # 1 row, 3 columns  
axes[0,0].plot(x, x*2, color='red')
axes[0,0].set_title('x*2')

axes[0,1].plot(x, x**2, color='blue')
axes[0,1].set_title('x**2')

axes[1,0].plot(x, x**3, color='green')
axes[1,0].set_title('x**3')

axes[1,1].plot(x, x**4, color='orange')
axes[1,1].set_title('x**4')

axes[0,2].bar(x, x**2, color='blue')
axes[0,2].set_title('Bar x**2')

axes[1,2].scatter(x, x**4, color='orange')
axes[1,2].set_title('Scatter x**4')

plt.tight_layout()
plt.show()
```
