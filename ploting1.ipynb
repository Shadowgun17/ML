#### Key Terms
1. Data Point :- (x,y) value pair <br> <br>
2. X-axis / Y-axis :- horizontal line(input) / vertical line(output) <br> <br>
3. Figure :- Entire Page or Canvas or Paper / top-level container for all elements of plot. Single figure can hold one or multiple Axes <br> <br>
4. Axes(subplot) :- Graph box / actual region where data is plotted. Contains x-axis, y-axis, ticks, labels, title and plot itself. <br> <br>
5. Plot :- visual representaion or graph created  <br> <br>
6. Marker :- A symbol used to denote the actual data points in plot.  <br> <br>
7. Line Style :- Defines the pattern of the line connecting the data points.  <br><br>
8. Color :- The color is used for the lines, markers, bars, or other plot elements  <br><br>
9. Legend :-  A key that identifies different data series(lines, bars, etc.) shown on the plot.  <br><br>
10. Label :-  A descriptive name assigned to a specific data series. This name is used to create the Legend of the plot.<br> <br>
11. Title :- Text placed on top of axes to describe the plot.  <br> <br>
12. Grid :- A network of horizontal and vertical lines drawn across the plot area to aid in reading the values corresponding to the data points <br> <br>
13. Function :- A block of organized, reusable code that perform a single, related action  <br> <br>
14. Method :- A function that belongs to an object. <br> <br>
15. Parameter :- A named variable listed inside the parantheses in a function or method definition. It represents the input the function expects. <br> <br>
16. Keyword Argument :- An argument that is explicitly passed to a parameter using its name, typically in formate name=value. e.x plt.plot(x,y,**linestyle='--'** , **marker='o'**) where linestyle and merker are keyword arguments. <br> <br>
17. Object-Oriented API :- This is the preferred and most powerful way to use matplotlib. It involves explicitly creating and interacting with Figure and Axes objects. <br> <br>
18. DIP(Device Independent Pixel) :- A unit of measurement used in matplotlib to size elements like text, line thickness and figure size, independent of the screen resolution or the final output resolution. Ensures that elements appear to have the same visual size regardless of the output device.  <br> <br>
19. Backend :- Is the layer that handles the actual drawing and user interface interation. It determines how plot is rendered.
   Types:<br> 1) interactive: disply plots in dynamic window (e.g. TkAgg, Qt5Agg). <br>
              2) Non-interactive: creates static image files(PNG, JPG, PDF) without a display window. <br>  <br>

20. Artist :- Base class for everything visible on the figure. Everything you see on the matplotlib plot- figure, axes, title, lines, text and images and collections - is an Artist. <br><br>
21. DPI(Dots per Inch) :- controls resolution of graph. Higher DPI --> clear image

# pyplot API

### Univariant - Numerical

```python
data = {
    "Salary" : [25000, 30000, 37000, 28000, 39000,
                48000, 25000, 14000, 60000, 42000,
                23000, 19000, 44000, 78000, 66000,
                88000, 56000, 20000, 34000, 58000]
}

df = pd.DataFrame(data)
```

```python
# Line Plot
plt.plot(df['Salary'], color='blue', marker = "o", linestyle = '--', linewidth = '2')
plt.grid()
plt.show()
```

```python
# Histogram
plt.hist(df['Salary'], bins=5)
plt.show()
```

```python
# Box Plot
plt.boxplot(df['Salary'])
plt.show()
```

## Univariant - Categorial

```python
# pie chart
departments = ['HR','IT','Finance','Marketing','Customercare']
df['dept'] = np.random.choice(departments,  size=len(df))
```

```python
df.info()
```

```python
count = df['dept'].value_counts()
count
```

```python
plt.pie(count, labels=count.index, autopct='%1.2f', explode=(0,0.1,0,0.1,0))
plt.show()
```

```python
# Countplot
plt.bar(count.index, count, color=['blue', 'green', 'red', 'cyan', 'magenta'])
plt.show()
```

# Bivariant 

### Numerical

```python
age = [21,22,24,25,26,28,29,30,32,33,35,37,40,42,25,28,51,54,58,62]
sal = [35000,38500,42000,45000,48200,52000,55500,60000,64000,68500,72000,78000,85000,92000,105000,115000,128000,142000,155000,175000]

df.insert(0,'Age',age)
df['Salary'] = sal
print(df) 
```

```python
# Scatter plot
plt.scatter(df['Age'], df['Salary'], color='blue')
plt.show()
```

```python
# line plot
plt.plot(df['Age'],df['Salary'], color='black', marker='o', linestyle='--', linewidth='1')
plt.grid()
plt.show()
```

```python
# Bar chart
plt.bar(df['Age'],df['Salary'], color='blue', align='edge') # 'center' or 'edge'
plt.show()
```

## Bivariant: Numerical and Categorial

```python
hr_sal = df[df['dept'] == 'HR']['Salary']
hr_sal
```

```python
it_sal = df[df['dept'] =='IT']['Salary']
it_sal
```

```python
market_sal = df[df['dept'] =='Marketing']['Salary']
market_sal
```

```python
plt.boxplot([hr_sal,it_sal,market_sal], tick_labels=['HR','IT','Marketing'])
plt.show()
```

```python
# Pie chart
salary_by_dept = df.groupby('dept')['Salary'].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept, labels= salary_by_dept.index, autopct='%1.2f')
plt.show()
```

```python
# Bar plot
hr_mean = sum(hr_sal)/len(hr_sal)
it_mean = sum(it_sal)/len(it_sal)
market_mean = sum(market_sal)/len(market_sal)
```

```python
plt.bar(['HR','IT','Marketing'], [hr_mean, it_mean, market_mean], color=['blue','green','red'])
plt.show()
```
