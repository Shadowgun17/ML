# Encoding
<h5> Nominal -> One Hot Encoding <br>
    Ordinal -> Ordinal Encoding, Label Encoding<br> </h5>

<h5> When to use ? <br>
If transforming output(y), Label Encoder<br>
If transforming input(X), Ordinal Encoder(for ordered data) or One-Hot Encoder(for unordered data)</h5>

# Scaling <br>

<h4> 1> Standardization (Z score Normalization) <br>
2> Normalization -> MinMax Scaling, Mean Normalization(Mean Centering), Max Absolute Scaling, Robust Scaling <br><br>

<h4> Standardization: <br> xi' =  (xi - x_mean)/ standard_deviation  [mean=0, std=1]</h4> <br>

<h4> Normalization: <br>
MinMax Scaling xi' = (xi - x_min)/(x_max - x_min)  [range 0 to 1]<br>
Mean Normalization xi' = (xi - x_mean)/(x_max - x_min) [range -1 to 1]<br>
Max Absolute Scaling xi' = xi / |x_max|<br>
Robust Scaling xi' = (xi - x_median)/IQR<br>
<h4> <br>

<h2> When to use? <br> </h2>
<h4> Standardization : K-means(Euclidean distance), K-Nearest-Neighbour(Euclidean distance), Principle Component Analysis(PCA), Artificial Neural Network, Gradient Descent <br> <br>
Normalization : <br>
MinMax Scaling -> when Min and Max of given data is known <br>
Mean Normalization -> Centered Data is required, alternative of Mean Normalization: Standardization <br>
Mean Absolute Scaling -> Sparce Data(data with many zeroes) <br>
Robust Scaling -> Robust to Outliers
</h4>
