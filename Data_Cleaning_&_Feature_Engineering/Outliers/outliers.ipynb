```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.boxplot(data=tips, x='total_bill', y='day', hue='sex')
```

```python
# Removing outliers
percentile25_1 = tips['tip'].quantile(0.25)
percentile75_1 = tips['tip'].quantile(0.75)
IQR = percentile75_1 - percentile25_1
lower_bound1 = percentile25_1 - (1.5 * IQR)
upper_bound1 = percentile75_1 + (1.5 * IQR)
```

```python
percentile25_2 = tips['total_bill'].quantile(0.25)
percentile75_2 = tips['total_bill'].quantile(0.75)
IQR = percentile75_2 - percentile25_2
lower_bound2 = percentile25_2 - (1.5 * IQR)
upper_bound2 = percentile75_2 + (1.5 * IQR)
```

```python
print(tips[(tips['tip'] < lower_bound1) | (tips['tip'] > upper_bound1)])
```

```python
print(tips[(tips['total_bill'] < lower_bound2) | (tips['total_bill'] > upper_bound2)])
```

```python
new_tips = tips[(tips['tip'] > lower_bound1) & (tips['tip'] < upper_bound1)]
new_total_bill = tips[(tips['total_bill'] > lower_bound2) & (tips['total_bill'] < upper_bound2)]
```

```python
plt.subplot(2,2,1)
plt.title('tips Without Outliers')
sns.boxplot(data=new_tips, x='tip', y='day')

plt.subplot(2,2,2)
plt.title('tips With Outliers')
sns.boxplot(data=tips, x='tip', y='day')

plt.subplot(2,2,3)
plt.title('total_bill Without Outliers')
sns.boxplot(data=new_total_bill, x='total_bill', y='day')

plt.subplot(2,2,4)
plt.title('total_bill With Outliers')
sns.boxplot(data=tips, x='total_bill', y='day')

plt.tight_layout()
```

```python
# Caping outliers
df = pd.DataFrame(tips)
df['tip'] = np.where(df['tip'] > upper_bound1, upper_bound1,
                     np.where(df['tip'] < lower_bound1, lower_bound1, df['tip']))
```

```python
sns.boxplot(data=df, x='tip', y='day')
```

```python
df['total_bill'] = np.where(df['total_bill'] > upper_bound2, upper_bound2,
                     np.where(df['total_bill'] < lower_bound2, lower_bound2, df['total_bill']))
```

```python
sns.boxplot(data=df, x='tip', y='day')
```
