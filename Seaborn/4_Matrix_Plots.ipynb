```python
import seaborn as sns
import plotly as px
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
planets = sns.load_dataset('planets')
gap = px.data.gapminder()
```

# Matrix Plot

### Heat Map
Plot rectangular data as a color-coded matrix


<p> The pivot function is a data transformation tool used to reorganize, summarize, and reshape data. It essentially turns "long-form" data (where variables are stored in rows) into "wide-form" data (where variables become column headers).
</p>

<pre>

pivot(): Used when you have unique index/column pairs. It will throw an error if there are duplicate 
entries for a single cell.

pivot_table(): The "heavy lifter." It can handle duplicate entries by aggregating them 
(e.g., taking the mean, sum, or max of multiple temperature readings for the same hour).

</pre>

<pre>
Figure Level Function: None
Axis Level Function: heatmap, clustermap
</pre>

```python
heat_df = gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
# ax lvl fun -> no fig lvl fun
plt.figure(figsize=(15,10))
sns.heatmap(data=heat_df)
```

```python
gap[gap['continent'] == 'Europe'].pivot(index='country', columns='year', values='lifeExp')
```

```python
plt.figure(figsize=(15,10))
heat_df = gap[gap['continent'] == 'Europe'].pivot(index='country', columns='year', values='lifeExp')
sns.heatmap(heat_df, annot=True, linewidths=0.5, cmap='winter')
```

```python
sns.heatmap(titanic.corr(numeric_only=True), cmap='icefire', annot=True) # cmap = 'YlGnBu' or 'coolwarm' or 'icefire'
```

```python
import pandas as pd

df = pd.DataFrame({
    "Maths":[90,80,70,60],
    "Science":[85,95,65,75],
    "English":[70,60,80,90]
})

sns.heatmap(df.corr(), annot=True, cmap="YlGnBu") 
```
