```python
import seaborn as sns
import plotly as px
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
planets = sns.load_dataset('planets')
gap = px.data.gapminder()
```

## Relation Plot

<pre>
> To see statistical realtion between two or more variables.
> Bivariant Analysis

Plot under this Section:
> scatter plot
> line plot
> facet plot
</pre>

<pre> 
Figure lvl fun : relplot
Axis lvl fun : scatterplot, lineplot
</pre>

### Scatter plot

```python
# Axis Level Function
sns.scatterplot(data=tips, x='tip', y='total_bill', hue='sex', size='size', style='time' , palette='YlGnBu')
```

```python
# Figure Level Function
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter')  
```

### Line plot

```python
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
# Axis level
sns.lineplot(data=temp_df, x='year', y='lifeExp')
```

```python
# Figure level
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
```

```python
target_df = gap[gap['country'].isin(['India','Pakistan','China'])]
```

```python
sns.lineplot(data=target_df, x='year', y='lifeExp', hue='country')
```

```python
sns.relplot(data=target_df, x='year', y='lifeExp', kind='line', hue='country')
```

```python
target_df2 = gap[gap['country'].isin(['India','Brazil','Germany'])]
```

```python
sns.relplot(data=target_df2, x='year', y='lifeExp', kind='line', hue='country', style='continent')
```

```python
target = {'country':['India'],'year':[2007,2002]}
ans = gap[gap[['country','year']].isin(target).all(axis=1)]
print(ans['country'].to_numpy())
```

### Facet plot
<p> Only works with relplot(fig-level function)

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='smoker', row='sex' ) # use 'col' or 'row' instead of 'hue'
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='line', col='sex', row='day' ) 
```

```python
# Column Wrap
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=4)
```
