# Type of function
<pre> 
1> Figure level
2> Axis level
</pre>

# Classification
<pre>   
1> Relational Plot(visualizing trends) :- Scatter plot, Line plot 

2> Distribution Plot(visualizing spread) :- Histogram, kde plot, Rug plot

3> Categorical Plot(comparing groups) :- categorical estimate plots[Bar plot, Count plot, Point plot], 
                                         categorical distribution plots[Box plot, Violin plot], 
                                         categorical scatter plots[Swarm plot, Strip plot]

4> Matrix Plot(complex correlation) :- Heat map, Cluster map

5> Regression Plot(predictive trends) :- Regression plot, Residual plot

6> Multi Plot :- Join plot, Pair plot
</pre>

## Analysis 
<pre>
> Univariant (one variable) (Distribution/Counts) :- Distribution plots, Categorical plots
> Bivariant (two variable) (Relationships/Comparisons) :- Relation plots, Regression plots, Categorical plots
> Multivariant (three+ variables) (Patterns/Groups) :- Multi plot
</pre>

# palette vs cmap
<pre>
palette: Used for discrete or categorical mapping (e.g., "Male" vs "Female"). 
It maps the categories in your hue column to specific colors.

cmap: Used for continuous numerical mapping (e.g., mapping a temperature range of 0–100 to a gradient).
</pre>

# cmap arguments
<pre>
'viridis' (Default): Perceptually uniform (great for accessibility/colorblindness).

'plasma', 'inferno', 'magma': High-contrast, dark-to-bright options.

'Greens', 'Blues', 'Reds': Simple single-color gradients.

'YlGnBu': Yellow to Green to Blue (very popular for readability).

'summer': green --> yellow

'autumn': Red --> Yellow

'winter': Blue --> Green

'spring': Magenta --> Yellow

> Reversing: You can append _r to any of the names above (e.g., 'summer_r', 'viridis_r', 'coolwarm_r') to flip the color scale.
</pre>
