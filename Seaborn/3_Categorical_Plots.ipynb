Figure Level Function: catplot

## 1> Categorical Scatter Plot

### Strip Plot

```python
sns.stripplot(data=tips, x='day', y='total_bill')
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', hue='day')
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', hue='day', jitter=False)
```

```python
day_colors = {
    "Thur": "#3498db",  # Soft Blue
    "Fri":  "#e74c3c",  # Soft Red
    "Sat":  "#f1c40f",  # Bright Yellow
    "Sun":  "#2ecc71"   # Emerald Green
}

sns.stripplot(data=tips, x='day', y='total_bill', hue='day', palette=day_colors)
```

```python
# Fig lvl fun
sns.catplot(data=tips, x='day',y='total_bill', kind='strip', hue='day', jitter=.15)
```

### Swarm Plot

```python
# Fig Lvl Fun
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm', hue='day')
```

```python
# Ax Lvl Fun
sns.swarmplot(data=tips, x='day', y='total_bill', hue='day')
```

## 2> Categorical Distribution Plot

### Box Plot
<pre>  
A Boxplot is a standardized way of displaying the distribution of data based on a five number summary
('minimum',first quartile[q1], median, third quartile[q3] and 'maximum'). 

It can tell you about outliers and what their values are. 

Boxplot can also tell you if your data is symmetrical, how tightly your data is grouped and 
if how your data is skewed.
</pre>

```python
sns.boxplot(data=tips, x='day', y='total_bill')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='box', hue='sex')
```

```python
# single boxplot
sns.boxenplot(data=tips, y='total_bill')
```

### Violin Plot (Boxplot + KDE Plot)

```python
sns.violinplot(data=tips, x='day', y='total_bill')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', hue='sex', height=6, aspect=2)
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', hue='sex', split=True)
```

## 3> Categorical Estimate Plot

### Bar Plot
<pre>
When there are multiple observations in each category, it also uses bootstrapping to compute 
confidence interval around the estimate, which is plotted using error bars.
</pre>

```python
sns.barplot(data=tips, x='sex', y='total_bill', hue='sex', errorbar=None)
```

```python
import numpy as np
sns.catplot(data=tips, x='sex', y='total_bill', kind='bar' , hue='smoker', errorbar=None, estimator=np.mean)
```

### Point Plot

```python
sns.catplot(data=tips, x='sex', y='total_bill', kind='point')
```

```python
sns.pointplot(data=tips, x='sex', y='total_bill', hue='smoker', estimator=np.min)
```

```python
sns.pointplot(data=tips, x='sex', y='total_bill', hue='smoker', errorbar=None)
```

### Count Plot
<pre>  
A special case for bar plot is when you want to show the number of observations in each category rather
than computing a statistic for a second variable. This is similar to a histogram over a categorical,
rather than quantitative variable.
</pre>

```python
sns.countplot(data=tips, x='sex')
```

```python
sns.countplot(data=tips, x='sex', hue='day')
```

```python
# Faceting using catplot
sns.catplot(data=tips, x='sex', y='total_bill', kind='box', col='smoker', row='time')
```
