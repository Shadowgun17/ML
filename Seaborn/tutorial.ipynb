### Histogram

```python
sns.histplot(data = tips, x='total_bill')
```

```python
# Fig lvl fun
sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
sns.displot(data=tips, x='tip', kind='hist', hue='sex', element='step')
```

### KDE Plot
<pre>
Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian Kernel,
producing a continuous density estimate.
</pre>

```python
tips
```

```python
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='kde', hue='sex', fill=True)
```

### Rug Plot
<pre>
Plot marginal distributions by drawing ticks along x and y axes.

This function is intended to complement other plots by showing the location of individaul oberservations 
in an unobtrusive way.
</pre>

```python
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
```

```python
# Bi-variate histogram
sns.histplot(data=tips, x='total_bill', y='tip')
```

# Matrix Plot

### Heat Map
<pre>  
Plot rectangular data as a color-coded matrix
</pre>

<p> The pivot function is a data transformation tool used to reorganize, summarize, and reshape data. It essentially turns "long-form" data (where variables are stored in rows) into "wide-form" data (where variables become column headers).</p>

<pre>
pivot(): Used when you have unique index/column pairs. It will throw an error if there are duplicate entries for a single cell.

pivot_table(): The "heavy lifter." It can handle duplicate entries by aggregating them 
(e.g., taking the mean, sum, or max of multiple temperature readings for the same hour).
</pre>

```python
heat_df = gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
# ax lvl fun -> no fig lvl fun
plt.figure(figsize=(15,10))
sns.heatmap(data=heat_df)
```

```python
gap[gap['continent'] == 'Europe'].pivot(index='country', columns='year', values='lifeExp')
```

```python
plt.figure(figsize=(15,10))
heat_df = gap[gap['continent'] == 'Europe'].pivot(index='country', columns='year', values='lifeExp')
sns.heatmap(heat_df, annot=True, linewidths=0.5, cmap='winter') # cmap = viridis summer
```

```python
sns.heatmap(titanic.corr(numeric_only=True), cmap='icefire', annot=True) # cmap = 'YlGnBu' or 'coolwarm' or 'icefire'
```

```python
import pandas as pd

df = pd.DataFrame({
    "Maths":[90,80,70,60],
    "Science":[85,95,65,75],
    "English":[70,60,80,90]
})

sns.heatmap(df.corr(), annot=True, cmap="YlGnBu") 
```
