```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
planets = sns.load_dataset('planets')
gap = px.data.gapminder()
```

# Distribution Plot
<pre> 
> used for univariant analysis
> used to find the distribution
> range of the observation
> central tendency
> is the data bimodal?
> are there outliers?

plots under distribution plot:
> histplot
> kdeplot
> rugplot
</pre>

<pre>
Fig lvl fun: displot
Ax lvl fun: histplot, kdeplot, rugplot
</pre>

### Histogram

```python
sns.histplot(data = tips, x='total_bill')
```

```python
# Fig lvl fun
sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
# It's also possible to visualize the distribution of the categorical variable using the logic of a histogram
# Discrete bins are automatically set for categorical variables

sns.displot(data=tips, x='day', kind='hist')
```

```python
sns.displot(data=tips, x='tip', kind='hist', hue='sex', element='step')
```

```python
# Faceting using col and row -> not work with ax lvl funs 
sns.displot(data=titanic, x='age', kind='hist', col='sex', row='survived', element='step')
```

### KDE Plot
<pre>
Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian Kernel,
producing a continuous density estimate.
</pre>

```python
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='kde', hue='sex', fill=True)
```

### Rug Plot
<pre>
Plot marginal distributions by drawing ticks along x and y axes.

This function is intended to complement other plots by showing the location of individaul oberservations 
in an unobtrusive way.
</pre>

```python
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
```

```python
# Bi-variate histogram
sns.histplot(data=tips, x='total_bill', y='tip')
```

```python
# Bi-variate kdeplot
sns.kdeplot(data=tips, x='total_bill', y='tip')
```
