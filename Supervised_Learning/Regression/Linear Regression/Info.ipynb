Residual Error : it is the vertical distance between an actual data point and the prediction line (the hyperplane) 
created by your model. It is the "mistake" the model made for that specific point.

If the residual is the error for a single data point, the Cost Function is the "Grand Total" of all those errors across your entire dataset.
In machine learning, the Cost Function (often denoted as $J(\theta)$) is a single number that tells you how well (or how poorly) your model 
is performing. A high cost means your model is guessing wildly; a low cost means your model is very accurate.

$$J(\theta_{0},\theta_{1}) = \frac{1}{2m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})^2$$

Where, 
$(h_\theta(x^{(i)}) - y^{(i)})$   is the residual (prediction - actual)
* **$J$**: Cost Function score.
* **$\theta_0, \theta_1$**: Model weights (Intercept and Slope).
* **$m$**: Total number of data points.
* **$\sum$**: Sum of all errors.
* **$h_\theta(x^{(i)})$**: Predicted value for the $i$-th point.
* **$y^{(i)}$**: Actual value for the $i$-th point.
* **$(...)^2$**: Squaring to handle negatives and penalize outliers.
