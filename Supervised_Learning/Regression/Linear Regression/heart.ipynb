```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("D://Machine Learning//sampledata//test//heart.csv")
df
```

# EDA

```python
df.info()
```

```python
df.describe()
```

```python
numeric_cols = ['Age','RestingBP','Cholesterol','MaxHR']

for i,col in enumerate(numeric_cols, 1):
    plt.subplot(2,2,i)
    sns.histplot(data=df, x=col, kde=True)
    plt.tight_layout()
```

```python
ch_mean = df.loc[df['Cholesterol'] != 0, 'Cholesterol'].mean()
ch_mean
```

```python
df['Cholesterol'] = df['Cholesterol'].replace(0, ch_mean)
df['Cholesterol'] = df['Cholesterol'].round(2)
```

```python
Rbp_mean = df.loc[df['RestingBP'] != 0, 'RestingBP'].mean()
df['RestingBP'] = df['RestingBP'].replace(0, Rbp_mean)
df['RestingBP'] = df['RestingBP'].round(2)
```

```python
numeric_cols = ['Age','RestingBP','Cholesterol','MaxHR']

for i,col in enumerate(numeric_cols, 1):
    plt.subplot(2,2,i)
    sns.histplot(data=df, x=col, kde=True)
    plt.tight_layout()
```

```python
df = pd.get_dummies(df, drop_first=True, dtype='int64')
```
