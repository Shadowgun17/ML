# Decision Tree

At its core, a **Decision Tree** is a supervised learning algorithm used for both classification and regression. It mimics human decision-making by breaking down a complex dataset into smaller, more manageable subsets based on specific features.

Think of it as an inverted tree: it starts with a single "root" and branches out into various "leaves."

---

## 1. How a Decision Tree Works

The algorithm works by repeatedly "splitting" the data based on the attribute that provides the most information. It asks a series of True/False questions until it reaches a final conclusion.

* **Root Node:** Represents the entire dataset and the first decision point.
* **Decision Nodes (Internal Nodes):** Points where the data is split into branches based on a specific feature.
* **Leaf Nodes (Terminal Nodes):** The final output or "class" (e.g., "Yes" or "No"). No further splitting occurs here.

---

## 2. The Logic Behind the Split

The tree doesn't split randomly. It uses mathematical criteria to ensure that the resulting groups are as "pure" as possible. The two most common methods are:

1. **Gini Impurity:** Measures the probability of a random element being incorrectly classified. A Gini score of 0 means the node is "pure" (all elements belong to one class).
2. **Information Gain (Entropy):** Measures the amount of "disorder" or uncertainty. The algorithm chooses the split that reduces entropy the most.

---

## 3. Advantages and Disadvantages

| **Pros** | **Cons** |
| --- | --- |
| **Easy to Interpret:** You can literally visualize the logic. | **Overfitting:** Trees can become too complex and "memorize" noise in the data. |
| **No Scaling Needed:** Works fine with non-normalized data. | **Instability:** A small change in the data can result in a completely different tree. |
| **Handles Mixed Data:** Can process both numerical and categorical features. | **Bias:** Can be biased toward features with more levels/categories. |

---

## 4. Preventing Overfitting (Pruning)

Because decision trees tend to grow until they perfectly fit the training data (which makes them perform poorly on new data), we use a technique called **Pruning**.

* **Pre-pruning:** Setting a `max_depth` (the maximum levels the tree can grow) or a `min_samples_split` during training.
* **Post-pruning:** Allowing the tree to grow fully and then removing branches that provide little power.





# Choosing the Node

Choosing the root node is the most critical step in building a decision tree. The goal is to select the feature that best partitions the data into the most "pure" or homogenous groups.

To do this, the algorithm calculates the **Impurity** of each feature and picks the one that reduces uncertainty the most.

---

## 1. The Concept of Purity

A node is considered **pure** if all the data points in that node belong to the same class. If a node has a 50/50 mix of classes, it is considered **impure**.

The "Root Node" is the feature that results in the highest "Information Gain" or the lowest "Gini Impurity" after the first split.

---

## 2. Common Mathematical Criteria

### A. Gini Impurity (Used by CART)

Gini Impurity measures the probability of a random element being incorrectly classified. The formula for a node $t$ is:

$$Gini(t) = 1 - \sum_{i=1}^{c} (P_i)^2$$

* **$P_i$**: The probability of an object being classified into a particular class.
* **Goal**: Pick the feature with the **lowest** weighted Gini Impurity for the resulting child nodes.

### B. Information Gain & Entropy (Used by ID3 / C4.5)

Entropy measures the "disorder" or randomness in the data.

$$Entropy(S) = - \sum_{i=1}^{c} P_i \log_2(P_i)$$

**Information Gain** is the difference between the entropy of the parent node and the weighted average entropy of the child nodes.

* **Goal**: Pick the feature with the **highest** Information Gain.

---

## 3. Step-by-Step Selection Process

1. **Calculate the Total Impurity:** Find the Entropy or Gini Impurity of the entire dataset before any splits.
2. **Test Every Feature:** For every available feature (e.g., "Outlook", "Temperature", "Humidity"):
* Split the data based on that feature.
* Calculate the weighted impurity of the resulting branches.


3. **Calculate the Gain:** Subtract the new weighted impurity from the original total impurity.
4. **Select the Winner:** The feature that provides the **largest reduction in impurity** becomes the Root Node.

---

## Example Scenario: Playing Tennis

Imagine you are deciding whether to play tennis based on the weather.

| Outlook | Humidity | Play? |
| --- | --- | --- |
| Sunny | High | No |
| Overcast | High | Yes |
| Rain | Normal | Yes |

If splitting by **Outlook** results in one branch being 100% "Yes" (Overcast), while splitting by **Humidity** results in two branches that are still a 50/50 mix, the algorithm will choose **Outlook** as the root node because it simplified the classification more effectively.




# => Calculating the entropy

Entropy is a measure of the "disorder" or unpredictability of the data in a given node. When we talk about a node with two possible outcomes (a binary classification), the formula simplifies into a very specific form.

The Formula for Two Classes (Binary)

If a node only has two possible outcomes—let's call them **Positive (+)** and **Negative (-)**—the formula expands to:

$$H(S) = -P_{(+)} \log_2(P_{(+)}) - P_{(-)} \log_2(P_{(-)})$$

#### Key Values to Remember:

* **Pure Node ($H = 0$):** If the node contains only one class (e.g., all 10 samples are "Positive"), then $P_{(+)} = 1$ and $P_{(-)} = 0$. The entropy is **0**. There is no disorder.
* **Maximum Disorder ($H = 1$):** If the node is a perfect 50/50 split (e.g., 5 "Positive" and 5 "Negative"), then $P_{(+)} = 0.5$ and $P_{(-)} = 0.5$. The entropy is **1**. This is the highest level of uncertainty.

---

Calculating "Weighted Entropy" for Two Child Nodes

When you split a parent node into two child nodes (Node A and Node B), you need to calculate the **Weighted Average Entropy** of the children to see if the split was actually useful.

$$\text{Weighted Entropy} = \left( \frac{N_A}{N_{total}} \times H(A) \right) + \left( \frac{N_B}{N_{total}} \times H(B) \right)$$

* **$N_A, N_B$**: Number of samples in Node A and Node B.
* **$N_{total}$**: Total samples in the parent node.
* **$H(A), H(B)$**: The entropy of each individual child node.

---

### 4. Practical Example

Imagine a parent node with **10 samples** (6 "Yes", 4 "No"). You split it into two nodes:

* **Node A:** 4 samples (4 "Yes", 0 "No") $\rightarrow$ $H(A) = 0$ (Pure)
* **Node B:** 6 samples (2 "Yes", 4 "No") $\rightarrow$ $H(B) = -(\frac{2}{6}\log_2\frac{2}{6}) - (\frac{4}{6}\log_2\frac{4}{6}) \approx 0.918$

**The Weighted Entropy of the split is:**


$$\left( \frac{4}{10} \times 0 \right) + \left( \frac{6}{10} \times 0.918 \right) = 0.5508$$

Since the original parent entropy was approximately $0.971$, and the new weighted entropy is $0.551$, the **Information Gain** is $0.971 - 0.551 = 0.420$. This indicates a good split!

# => Choosing Root Node

To choose the root node, you calculate the **Information Gain** for every feature in your dataset. The feature that provides the highest Information Gain is selected as the root.

Think of Information Gain as the reduction in "messiness" (Entropy) after you split the data.

---

### Step 1: Calculate the Entropy of the Entire Dataset (Parent)

Before you split anything, find the entropy of your target variable (the "Label").

**Formula:**


$$H(S) = - \sum_{i=1}^{c} P_i \log_2(P_i)$$

**Example:** Suppose you have 10 rows of data. 6 are "Yes" and 4 are "No."

* $P_{Yes} = 6/10 = 0.6$
* $P_{No} = 4/10 = 0.4$
* $H(S) = -(0.6 \log_2 0.6) - (0.4 \log_2 0.4) \approx \mathbf{0.971}$

---

### Step 2: Calculate the Entropy for Each Feature

Pick a feature (e.g., "Windy") and split the data into groups based on its values (e.g., "True" and "False"). Calculate the entropy for each resulting group.

**Example: Splitting by "Windy"**

* **Group 1 (Windy = True):** 4 rows (2 Yes, 2 No) $\rightarrow$ $H(True) = 1.0$ (Perfectly messy)
* **Group 2 (Windy = False):** 6 rows (4 Yes, 2 No) $\rightarrow$ $H(False) \approx 0.918$

---

### Step 3: Calculate the Weighted Average Entropy

You can't just look at the child nodes separately; you must weigh them by how many samples they contain.

$$\text{Weighted Entropy} = \sum \left( \frac{\text{Samples in Child}}{\text{Total Samples}} \times H(\text{Child}) \right)$$

**Calculation for "Windy":**

* $(4/10 \times 1.0) + (6/10 \times 0.918) = 0.4 + 0.550 = \mathbf{0.950}$

---

### Step 4: Calculate Information Gain

Information Gain is the difference between the original entropy and the new weighted entropy.

$$\text{Information Gain} = H(\text{Parent}) - \text{Weighted Entropy}(\text{Children})$$

**Calculation for "Windy":**

* $0.971 - 0.950 = \mathbf{0.021}$

---

### Step 5: Compare and Choose

Repeat Steps 2 through 4 for every other feature in your dataset (e.g., "Outlook," "Humidity," "Temperature").

| Feature | Information Gain |
| --- | --- |
| **Outlook** | **0.246** |
| Humidity | 0.151 |
| Windy | 0.021 |

**The Result:** Since **Outlook** has the highest Information Gain, it becomes the **Root Node**. All other features will be tested again at the next level of the tree.
