Support Vector Machine (SVM) is a powerful supervised machine learning algorithm used for both classification and regression, though it is most commonly applied to classification problems.

The core objective of SVM is to find a **hyperplane** in an $N$-dimensional space (where $N$ is the number of features) that distinctly classifies the data points.

---

## Core Terms in SVM

* **Hyperplane:** A decision boundary that separates different classes. In 2D, this is a line; in 3D, it is a plane.
* **Support Vectors:** The data points that are closest to the hyperplane. These points are critical because if they are moved, the position of the decision boundary changes.
* **Margin:** The distance between the hyperplane and the nearest data point from either class. SVM aims to **maximize** this margin.
* **Kernel:** A mathematical function used to transform low-dimensional input space into a higher-dimensional space to make non-linear data linearly separable.
* **Hard Margin:** Used when data is perfectly linearly separable; it allows no misclassifications.
* **Soft Margin:** Used when data is noisy; it allows some misclassifications to achieve a better overall fit (controlled by a "C" parameter).

---

## The Mathematical Formula

The goal is to find the vector $w$ (weights) and scalar $b$ (bias) for the hyperplane defined by the equation:

$$w \cdot x + b = 0$$

For any data point $x_i$, the classifier function $f(x_i)$ is:

$$y_i(w \cdot x_i + b) \ge 1$$

Where:

* $y_i$ is the class label (+1 or -1).
* $w$ is the weight vector normal to the hyperplane.
* $b$ is the bias.

To maximize the margin, we minimize the following objective function:

$$\min \frac{1}{2} \|w\|^2$$

Subject to the constraint: $y_i(w \cdot x_i + b) \ge 1$ for all $i$.

---

## Steps for Calculation

Calculating an SVM model manually involves solving a quadratic programming problem. Here is the conceptual workflow:

1. **Define the Input:** Start with a training dataset of $n$ points of the form $(x_i, y_i)$.
2. **Select the Kernel:** Decide if the data is linear. If not, choose a kernel (like RBF or Polynomial) to map data to a higher dimension.
3. **Set Constraints:** For each data point, ensure it lies on the correct side of the margin: $y_i(w \cdot x_i + b) \ge 1$.
4. **Optimization:** Use Lagrange Multipliers to transform the constrained optimization problem into an unconstrained one. This involves finding values for $\alpha$ (Lagrangian multipliers) that maximize:

$$L_d = \sum \alpha_i - \frac{1}{2} \sum \sum \alpha_i \alpha_j y_i y_j (x_i \cdot x_j)$$


5. **Calculate Weights and Bias:** Once the optimal $\alpha$ values are found, the weight vector $w$ is calculated as $w = \sum \alpha_i y_i x_i$. The bias $b$ is solved using any support vector where the equality $y_i(w \cdot x_i + b) = 1$ holds.
6. **Prediction:** New points are classified based on which side of the hyperplane they fall: $\text{sign}(w \cdot x_{new} + b)$.
