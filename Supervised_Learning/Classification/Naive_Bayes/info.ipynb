Naive Bayes is a supervised learning algorithm used for classification. It is based on **Bayes' Theorem** and is called "Naive" because it makes the simplifying assumption that all input features are independent of each other—an assumption that is rarely true in real-world data but makes the algorithm incredibly fast and effective.

### 1. The Mathematical Foundation

The algorithm calculates the probability of a hypothesis (a class label) given the evidence (the features). This is expressed by the formula:

$$P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}$$

In the context of Machine Learning, we can rewrite this as:

* **$P(c|x)$ (Posterior Probability):** The probability of class $c$ given predictor $x$.
* **$P(x|c)$ (Likelihood):** The probability of predictor $x$ given class $c$.
* **$P(c)$ (Prior Probability):** The probability of class $c$ occurring generally.
* **$P(x)$ (Evidence):** The probability of the predictor $x$ occurring.

---

### 2. Why is it "Naive"?

In a complex dataset, features usually influence one another. For example, if you are identifying a fruit, "roundness" and "red color" might be correlated. Naive Bayes ignores these correlations and treats each feature as if it contributes independently to the probability that the fruit is an apple.

Despite this "naive" oversimplification, it often performs surprisingly well, especially for **Natural Language Processing (NLP)** tasks like spam detection.

---

### 3. Common Variants

Depending on the distribution of your data, you will choose a specific type of Naive Bayes:

| Type | Best Used For... |
| --- | --- |
| **Gaussian** | Continuous data that follows a normal (bell curve) distribution. |
| **Multinomial** | Discrete data, typically word counts for text classification. |
| **Bernoulli** | Binary/Boolean features (e.g., "word is present" vs. "word is absent"). |

---

### 4. Pros and Cons

**Advantages:**

* **Speed:** It is much faster than more complex algorithms like Support Vector Machines or Deep Learning.
* **Scalability:** It handles high-dimensional data (many features) very well.
* **Small Datasets:** It requires less training data than other models to reach reasonable accuracy.

**Disadvantages:**

* **Zero Frequency:** If a category in the test data was never seen in training, the model assigns a 0% probability. This is usually fixed using "Laplace Smoothing."
* **Independence Assumption:** If features are highly correlated, the model's performance may suffer.

---

### 5. Common Use Cases

* **Spam Filtering:** Identifying if an email is "Spam" or "Ham" based on word frequency.
* **Sentiment Analysis:** Classifying social media posts as Positive, Negative, or Neutral.
* **Recommendation Systems:** Predicting whether a user will like a product based on their browsing history.

# Conditional Probability

Conditional probability is the probability of an event occurring given that another event has already occurred. 
It answers the question: "What is the probability of A happening, given that B has already happened?" 

## Definition

Conditional probability is denoted as **P(A|B)** and is read as "the probability of A given B."

Mathematically, it is defined as:

$$P(A|B) = \frac{P(A \cap B)}{P(B)}$$

Where:
* **P(A|B)** = Conditional probability of A given B
* **P(A ∩ B)** = Joint probability of both A and B occurring
* **P(B)** = Probability of event B occurring

**Important:** P(B) must be greater than 0, otherwise the conditional probability is undefined.

## Intuition

Instead of looking at all possible outcomes, conditional probability restricts the sample space to only those outcomes where event B has occurred. Then, we calculate what fraction of those outcomes also include event A.

## Example

Suppose we have a deck of 52 playing cards:
* **Event A:** Drawing a King
* **Event B:** Drawing a Red card

What is P(A|B) — the probability of drawing a King given that we've drawn a Red card?

* There are 26 red cards in total
* There are 2 red Kings (King of Hearts, King of Diamonds)
* Therefore: **P(King | Red) = 2/26 = 1/13 ≈ 0.077**

Without the condition, P(King) = 4/52 = 1/13. In this case, they happen to be equal, but this is not always the case.

## Key Properties

* **0 ≤ P(A|B) ≤ 1** — Conditional probability is always between 0 and 1
* **P(A|B) = P(A)** — If A and B are independent events
* **P(A|B) × P(B) = P(A ∩ B)** — The multiplication rule

# Derivation of Bayes Theorem

Bayes Theorem is derived from the definition of conditional probability.

Conditional probability is defined as:

$$P(A|B) = \frac{P(A \cap B)}{P(B)}$$

Similarly,

$$P(B|A) = \frac{P(A \cap B)}{P(A)}$$

From the second equation, we can solve for the joint probability:

$$P(A \cap B) = P(B|A) \cdot P(A)$$

Substitute this into the first equation:

$$P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}$$

This is Bayes Theorem, expressing the posterior probability \(P(A|B)\) in terms of the likelihood \(P(B|A)\), prior \(P(A)\), and evidence \(P(B)\).
