## Why to use it instead of a straight line?
<pre>
Linear regression tries to fit a straight line (y = mx + c) to the data.While great for predicting house prices,
it fails at classification for two main reasons:
1> Outliers: A single extreme data point can shift the entire straight line, leading to poor classification
             boundaries.
2> Invalid Probabilities: A straight line can predict values less than 0 or greater than 1. 

In classification,we need a probability—a value strictly between 0 and 1.
Logistic Regression solves this by "wrapping" the linear equation into an S-shaped curve 
(the Sigmoid function), which compresses any input into the range of (0, 1).
</pre>

The function $h_\theta(x)$ is the Hypothesis Function. It represents the model’s prediction for a given input $x$, parameterized by $\theta$. <br>

$h_\theta(x) = g(\theta_0 + \theta_1x_1)$<br>

$h_\theta(x)$ (The Hypothesis): This is the estimated probability that the output is $1$ for a given input $x$. <br>
                                In mathematical notation: $h_\theta(x) = P(y=1 | x; \theta)$.<br>
 
$g(z)$(The Sigmoid/Logistic Function): This is the "wrapper" function that squashes any input $z$ into a value between $0$ and $1$.

$ h_\theta(x) = \frac{1}{1 + e^{-(\theta_0 + \theta_1x_1)}} $ <br> 
where:<br>
g = $ \frac{1}{1 + e^{-z}} $ <br>
e = euler number(a mathematical constant, value = 2.71828) it smooths the curve <br>
z = $ \theta_0 + \theta_1x_1 $


## Loss Function
The Log Loss function (also known as Binary Cross-Entropy Loss) is the standard cost function used to measure the 
performance of a Logistic Regression model.<br><br>

Log Loss measures "uncertainty." It penalizes the model based on how far the predicted probability ($h_\theta(x)$) is from 
the actual label ($y$).<br>
If the actual label is 1 and the model predicts 0.99, the loss is very small.<br>
If the actual label is 1 and the model predicts 0.01, the loss is extremely high (tending toward infinity).

$J(\theta) = -\frac{1}{m} \sum_{i=1}^{m} [y^{(i)} \ln(h_\theta(x^{(i)})) + (1 - y^{(i)}) \ln(1 - h_\theta(x^{(i)}))]$  <br>

Where, <br>
m is number of samples <br>
$ y^{(i)} $ is actual label(0,1)<br>
$ h_\theta(x^{(i)}) $ predicted probability (0 - 1) <br>


Why the $(1-y)$ part?<br>
It acts as a "switch".<br>
When $y=1$, the second term $(1-1)$ becomes zero, leaving only the first term.<br>
When $y=0$, the first term becomes zero, leaving only the second term.

# Evaluation of Model

Evaluating a Logistic Regression model is different from evaluating a Linear Regression model. Since we are predicting **classes** (like "Success" vs. "Failure") rather than continuous numbers, we use metrics based on the **Confusion Matrix**.

---

## 1. The Confusion Matrix

The foundation of all evaluation metrics is a $2 \times 2$ table that compares the **Actual** values with the **Predicted** values.

* **True Positive (TP):** Predicted 1, Actual 1.
* **True Negative (TN):** Predicted 0, Actual 0.
* **False Positive (FP):** Predicted 1, Actual 0 (Type I Error).
* **False Negative (FN):** Predicted 0, Actual 1 (Type II Error).

---

## 2. Key Metrics

From the matrix above, we derive several percentages to see how the model "behaves."

### Accuracy

The overall percentage of correct predictions.


$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$

> **Note:** Accuracy can be misleading if your classes are imbalanced (e.g., 99% of emails are not spam).

### Precision

Of all the times the model predicted "Positive," how many were actually correct? Use this when **False Positives** are costly (e.g., marking a legitimate email as spam).


$$\text{Precision} = \frac{TP}{TP + FP}$$

### Recall (Sensitivity)

Of all the actual "Positive" cases, how many did the model catch? Use this when **False Negatives** are dangerous (e.g., missing a cancer diagnosis).


$$\text{Recall} = \frac{TP}{TP + FN}$$

### F1-Score

The "middle ground" (harmonic mean) between Precision and Recall. It’s the best metric when you have an uneven class distribution.


$$\text{F1} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}$$

---

## 3. Threshold Evaluation: ROC and AUC

Logistic Regression outputs a **probability**. By default, the threshold is 0.5, but we can change it.

* **ROC Curve (Receiver Operating Characteristic):** A plot of the **True Positive Rate** vs. the **False Positive Rate** at every possible threshold.
* **AUC (Area Under the Curve):** A single number from 0 to 1 representing model quality.
* **1.0:** Perfect model.
* **0.5:** No better than a random guess (like flipping a coin).



---

## 4. Summary Checklist for Evaluation

When you are building your models follow this hierarchy:

1. **Check the Confusion Matrix:** Get a "feel" for the errors.
2. **Look at F1-Score:** Especially if one class is much smaller than the other.
3. **Check AUC-ROC:** To see how well the model separates the two classes regardless of the threshold.
4. **Log Loss:** To see how "confident" the model's probabilities are (lower is better).
