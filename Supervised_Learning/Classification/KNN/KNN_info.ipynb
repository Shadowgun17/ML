K-Nearest Neighbors (KNN) is a simple, versatile, and "lazy" supervised learning algorithm used for both classification and regression. It operates on a very intuitive principle: similar things exist in close proximity


## How It Works

Instead of building a complex mathematical model during a training phase, KNN stores the entire dataset. When you give it a new data point to predict, it follows these steps:

1. **Calculate Distance:** It measures the distance between the new point and all other points in the dataset (most commonly using **Euclidean distance**).
2. **Find Neighbors:** It selects the $K$ number of points that are closest to the new point.
3. **Vote (Classification) or Average (Regression):**
* **Classification:** The new point is assigned to the class most common among its $K$ neighbors.
* **Regression:** The new point is assigned the average value of its $K$ neighbors.



---

## Key Components

### 1. Choosing "K"

The value of $K$ (the number of neighbors) is a hyperparameter that significantly impacts the model:

* **Small K (e.g., $K=1$):** The model is very sensitive to noise and outliers (overfitting).
* **Large K:** The model becomes smoother and more stable, but it might overlook local patterns (underfitting).
* **Tip:** It is often best to use an odd number for $K$ to avoid ties in classification.

### 2. Distance Metrics

The algorithm relies on calculating the "distance" between points in a multi-dimensional space. Common formulas include:

* **Euclidean Distance:** The straight-line distance between two points ($d = \sqrt{\sum (x_i - y_i)^2}$).
* **Manhattan Distance:** The sum of absolute differences ($d = \sum |x_i - y_i|$).
* **Minkowski Distance:** A generalized form of both Euclidean and Manhattan.

## Pros and Cons

| Advantages | Disadvantages |
| --- | --- |
| **Simple to understand:** Very intuitive logic. | **Slow on large data:** Since it calculates distances for every prediction, it’s computationally expensive. |
| **No Training Phase:** It’s a "lazy learner," so it doesn't need to be trained before making predictions. | **Memory Intensive:** Requires storing the entire dataset. |
| **Non-parametric:** Makes no assumptions about the underlying data distribution. | **Sensitive to Scale:** Features with larger ranges (e.g., Salary vs. Age) can dominate the distance calculation. |

---

## Practical Tip: Feature Scaling

Because KNN relies entirely on distance, you **must scale your data** (using techniques like Normalization or Standardization). If one feature has a range of 0–1 and another has a range of 0–1,000, the algorithm will treat the larger range as much more important, even if it isn't.
