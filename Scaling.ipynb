```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
 
```

```python
df = pd.read_csv("D://sampledata//campusx//Social_Network_Ads.csv")
```

```python
df
```

```python
df = df.iloc[:,2:]
```

# Train, Test, Split

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1),
                                                    df['Purchased'],
                                                    test_size=0.3,
                                                    random_state=0)

"""
df.drop('Purchased', axis=1): This creates X (the Features). It takes your entire table and removes 
                              the "Purchased" column. 
                              These are the variables used to make predictions (e.g., Age, Salary, Gender).

df['Purchased']: This creates y (the Target). This is the single column you want to predict.

test_size=0.3: This defines the ratio. 30% of your data will be set aside for testing, 
               and the remaining 70% will be used for training.

random_state=0: This is a "seed" for the random number generator. It ensures that every time you run this code, 
                you get the exact same split.
                This is crucial for debugging and making your results reproducible.

X_train: The input features the model uses to learn patterns.
y_train: The correct answers for the training set (the "labels").
X_test: New input features the model has never seen.
y_test: The "Answer Key" used to check if the model's predictions on X_test are correct.
"""
```

```python
X_train.shape, X_test.shape
```

#  Standardization StandardScaler  

<h4> It is technique to standardize the independent features present in the data in a fixed range </h4>

```python
scaler = StandardScaler()

# fit the scaler on the training data, it will learn the patterns
scaler.fit(X_train)

# transform both the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.mean_
```

```python
# Array form
X_train_scaled
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns)
```

```python
# Pandas data frame
X_train_scaled
```

```python
np.round(X_train.describe(), 1)

# np.round(..., 1) : This is a NumPy function applied to the entire table of statistics.
# It rounds every number in the summary to one decimal place.
# Without this, Pandas often displays numbers with 6+ decimal places (e.g., 25.333333), 
# which makes the table cluttered and harder to scan quickly.
```

```python
np.round(X_train_scaled.describe(), 1)
```

# Effect of Scaling

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12,5))
ax1.scatter(X_train['Age'], X_train['EstimatedSalary'], c = 'blue')
ax1.set_title('Before Scaling')

ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'], c = 'green')
ax2.set_title('After Scaling')

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize = (12,5))
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Age'], ax = ax1)
sns.kdeplot(X_train['EstimatedSalary'], ax = ax1)

ax2.set_title('After Scaling')
sns.kdeplot(X_train_scaled['Age'], ax = ax2)
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax = ax2)

plt.show()

# distribution remains same before and after standardization, but the scale changes
```

# Importance of Scaling

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
lr_scaled.fit(X_train_scaled, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Actual: ',accuracy_score(y_test, y_pred))
print('Scaled: ',accuracy_score(y_test, y_pred_scaled))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
dtc_scaled = DecisionTreeClassifier()
```

```python
dtc.fit(X_train, y_train)
dtc_scaled.fit(X_train_scaled, y_train)
```

```python
y_pred = dtc.predict(X_test)
y_pred_scaled = dtc_scaled.predict(X_test_scaled)
```

```python
print('Actual:',accuracy_score(y_test, y_pred))
print('Scaled:',accuracy_score(y_test, y_pred_scaled))
```

# Normalization

<h4> Types: <br>
MinMax Scaling, Mean Normalization, Max Absolute Scaling, Robust Scaling

## MinMax Scaling

```python
df = pd.read_csv("D://sampledata//campusx//wine_data.csv",header=None, usecols = [0,1,2])
df.columns = ['Class Label', 'Alcohol', 'Malic Acid']
```

```python
df
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop('Class Label', axis=1), 
                                                    df['Class Label'],
                                                    test_size=0.3,
                                                    random_state=0)

x_train.shape, x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(x_train)

x_train_scaled = scaler.transform(x_train) # array form
x_test_scaled = scaler.transform(x_test)
```

```python
# dataframe
x_train_scaled = pd.DataFrame(x_train_scaled, columns=x_train_scaled.columns)
x_test_scaled = pd.DataFrame(x_test_scaled, columns=x_train_scaled.columns)
```
