# ANOVA Test
<pre>  
It is used to compare means of more than two independent groups to check if at least one
group is significantly different.

Use ANOVA when you have one numerical variable and one categorical variable with 3+ levels.
</pre>

### ANOVA (Analysis of Variance) Formula

The **F-statistic** determines if the variation between group means is significantly larger than the variation within the groups.

$$F = \frac{MSB}{MSW}$$

**Where:**
* **$F$**: The F-ratio (The test statistic).
* **$MSB$ (Mean Square Between)**: The variance between the different groups (The **Signal**).
* **$MSW$ (Mean Square Within)**: The variance within the individual groups (The **Noise**).

---

### Components of the F-Statistic:
To find $MSB$ and $MSW$, we use:

$$MSB = \frac{SSB}{k - 1} \quad \text{and} \quad MSW = \frac{SSW}{N - k}$$

**Where:**
* **$SSB$**: Sum of Squares Between groups (how far group means are from the overall average).
* **$SSW$**: Sum of Squares Within groups (how much individual data points vary inside their own group).
* **$k$**: The number of groups (e.g., if testing 3 different ML models, $k=3$).
* **$N$**: The total number of all observations across all groups combined.
* **$k - 1$**: Degrees of freedom for the groups.
* **$N - k$**: Degrees of freedom for the error (within-group).

```python
import pandas as pd 
import seaborn as sns
from scipy.stats import  f_oneway
```

```python
df = sns.load_dataset('titanic')
df
```

```python
df = df[['age','pclass']].dropna()
```

```python
df['pclass'].unique()
```

```python
class1 = df[df['pclass'] == 1]['age']
class2 = df[df['pclass'] == 2]['age']
class3 = df[df['pclass'] == 3]['age']
```

```python
f_stats, p_val = f_oneway(class1, class2, class3)
```

```python
f_stats
```

```python
p_val
```

```python
alpha = 0.05
if p_val < alpha:
    print('Reject the null, there is significant difference between in atleast one passenger class')
else:
    print('Accept the null')
```
