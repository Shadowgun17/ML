# Covariance
Covariance measures how two variables change together. If one goes up while the other goes up, covariance is positive. If one goes up while the other goes down, it's negative.

### Sample Covariance Formula
Covariance ($Cov_{xy}$) measures the joint variability of two random variables.

$$Cov(x, y) = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{n - 1}$$

**Where:**
* **$Cov(x, y)$**: Covariance between variable $x$ and variable $y$.
* **$x_i, y_i$**: Individual data points in the sets $x$ and $y$.
* **$\bar{x}, \bar{y}$**: The arithmetic means (averages) of $x$ and $y$.
* **$n$**: The total number of data points.
* **$n - 1$**: Degrees of freedom (Bessel's correction for sample data).

Covariance just tells us the direction the relation is positive or negative but to actually see the strength of 
relation we have to use the correlation.

### Pearson Correlation Coefficient ($r$) Formula
Correlation measures the strength and direction of the linear relationship between two variables.

$$r = \frac{Cov(x, y)}{s_x \times s_y}$$

**Alternatively, the expanded version:**

$$r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}}$$

**Where:**
* **$r$**: The correlation coefficient (ranges from -1 to +1).
* **$Cov(x, y)$**: The covariance between $x$ and $y$.
* **$s_x, s_y$**: The standard deviations of $x$ and $y$ respectively.
* **$\bar{x}, \bar{y}$**: The means of $x$ and $y$.

It is always between -1 to 1.
