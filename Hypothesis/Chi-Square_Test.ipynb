# Chi-Square Test
<pre> 
It is used to check if there is significant association between two categorical variables.
</pre>

### Chi-Square ($\chi^2$) Formula
The Chi-Square test measures how expectations compare to actual observed data.

$$\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}$$

**Where:**
* **$\chi^2$**: The Chi-Square statistic.
* **$\sum$**: The summation symbol (add up the results for all categories).
* **$O_i$**: The **Observed** frequency (the actual count you measured).
* **$E_i$**: The **Expected** frequency (the count predicted by the Null Hypothesis).
* **$i$**: The specific category or "cell" in your table.

### Expected Frequency ($E_{ij}$) Formula
In a contingency table, the expected frequency for a specific cell is calculated as:

$$E_{ij} = \frac{T_r \times T_c}{N}$$

**Where:**
* **$E_{ij}$**: The **Expected Frequency** for the cell in row *i* and column *j*.
* **$T_r$**: The **Row Total** (the sum of all observations in that specific row).
* **$T_c$**: The **Column Total** (the sum of all observations in that specific column).
* **$N$**: The **Grand Total** (the total number of all observations in the entire table).

Degree of Freedom: $ df = (r - 1)(c - 1)  $<br>
where r = number of rows, c = number of columns

```python
import numpy as np 
import seaborn as sns
from scipy.stats import chi2_contingency
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
df
```

```python
df.info()
```

```python
contingency_table = pd.crosstab(df['sex'], df['survived'])
contingency_table
```

```python
chi2, p_val, dof, expected = chi2_contingency(contingency_table)
```

```python
expected
```

```python
dof
```

```python
p_val
```

```python
chi2
```

```python
alpha = 0.05
if p_val < alpha:
    print('Reject the null hypothesis, there is significant connection between gender and survival')
else:
    print("Accept the null hypothesis, there is no connection")
```
