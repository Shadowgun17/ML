# T Test
<pre>
> Unknown Population Standard Deviation <br>
> Small sample size usually less than 30 <br>
> Data points must be independent <br>
> Data is collected through random sampling <br>
</pre>

<pre>
1. The "Extra Level of Luck" (Uncertainty)

> Z-test: Assumes you know the Population Standard Deviation perfectly. You are only "guessing" the Mean.
> T-test: Acknowledges you are guessing two things at once: the Mean AND the Standard Deviation.
> The Effect: Because you are estimating the spread from a sample, the T-test is more cautious. It uses 
              "fatter tails" in its distribution to account for the risk that your sample's spread might be a fluke.

2. When the Difference Disappears (Convergence)

> Sample Size: As your n grows, your sample's standard deviation becomes an extremely accurate mirror of the true population.
> The Result: At high numbers, the "fat tails" of the T-distribution shrink. The T-distribution mathematically converges 
              into the Z-distribution.
> The Math: If you ran both tests on a million data points, the P-values would be identical to several decimal places.

3. Why we use the T-test by Default

> In almost all real-world research and Machine Learning, we neve truly know the population. Using a T-test 
  is the only way to be mathematically honest about our data.
> Universal Safety: The T-test is "self-correcting.
--> If n is small, it protects you from making bold claims (by being stricter).
--> If n is large, it automatically acts like a Z-test.


> Tools like Python (`scipy.stats`), R, and Excel use the T-test as the default because 
  it works in every scenario, whereas the Z-test only works in one specific (and rare) case.

</pre>

$$t = \frac{\bar{x} - \mu}{s / \sqrt{n}}$$

$$
\text{where: }
\begin{aligned}
t &= \text{Calculated T-score} \\
\bar{x} &= \text{Sample mean} \\
\mu &= \text{Hypothesized population mean} \\
s &= \text{Sample standard deviation} \\
n &= \text{Sample size}
\end{aligned}
$$

Degree of Freedom: $df = n - 1$

```python
import numpy as np
from scipy import stats
```

```python
sample = [172, 174, 168, 169, 171, 173, 175, 170, 169, 172]
sam_mean = np.mean(sample)
sam_std = np.std(sample,ddof=1)
n = len(sample)
pop_mean = 170
```

```python
t_test = (sam_mean - pop_mean)/(sam_std / np.sqrt(n))
t_test
```

```python
p_val = 2 * (1 - stats.t.cdf(abs(t_test),df = n-1))
p_val
```

```python
alpha = 0.05
if p_val < alpha:
    print('Reject the Null Hypothesis')
else:
    print("Fail to Reject the Null Hypothesis (Accept the null hypothesis)")
```
