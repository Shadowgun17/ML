# Z Test
<pre>
> Population >= 30  <br>
> Known Variance or Standard Deviation <br>
> Each data point must be independent of the others<br>
> if population < 30 and normally distributed, z-test can be applied 
</pre>

$$Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}$$

$$
\text{where: }
\begin{aligned}
\bar{x} &= \text{Sample mean} \\
\mu &= \text{Population mean} \\
\sigma &= \text{Standard deviation} \\
n &= \text{Sample size}
\end{aligned}
$$

```python
import numpy as np 
from scipy.stats import norm
```

```python
sample = [172, 174, 168, 169, 171, 173, 175, 170, 169, 172]
pop_mean = 170
pop_std = 3
sample_mean = np.mean(sample)
n = len(sample)
```

```python
z_score = (sample_mean  - pop_mean)/ (pop_std / np.sqrt(n))
z_score
```

```python
p_val = 2 * (1 - norm.cdf(abs(z_score)))
p_val
```

```python
alpha = 0.05

if p_val < alpha:
    print('Reject the null hypothesis')
else:
    print("Fail to Reject the null hypothesis(Accept the null)")
```
