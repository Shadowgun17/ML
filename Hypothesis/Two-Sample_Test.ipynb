# Two-Sample Test
<pre>
You use a Two-Sample Test when you want to compare the means (or proportions) of two distinct, 
independent groups to see if there is a statistically significant difference between them.
</pre>

--> Large sample (n>30) and population variance are known <br>
--> Two-Sample Z-Test: <br>
$$Z = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2}}}$$

**where:**
* $\bar{x}_1, \bar{x}_2$ = Means of group 1 and 2
* $\sigma_1, \sigma_2$ = Standard deviations of group 1 and 2
* $n_1, n_2$ = Sample sizes of group 1 and 2

--> Small sample or population variance are unknown <br>
--> Two-Sample T-Test: <br>
$$t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}$$

**where:**
* $\bar{x}_1, \bar{x}_2$ = Means of group 1 and 2
* $s_1, s_2$ = Standard deviations of group 1 and 2
* $n_1, n_2$ = Sample sizes of group 1 and 2

<br><br>
Degree of Freedom: $ df = \min(n_1 - 1, n_2 - 1)  $

```python
import numpy as np
from scipy import stats
```

```python
group1 = [85,88,90,92,87,85,89,91,86,88]
group2 = [82,84,80,83,81,79,78,85,84,83]
```

```python
t_stats, p_value = stats.ttest_ind(group1, group2, equal_var=False)
```

```python
t_stats, p_value
```

```python
alpha = 0.05
if p_value > alpha:
    print('Accept the Null')
else:
    print('Reject the Null')
```
