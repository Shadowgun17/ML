# What is Unsupervised Learning?

**Unsupervised learning** is a type of machine learning where the model is trained on data that has **no predefined labels or targets**. Unlike supervised learning, where the "answer key" is provided, the algorithm must find hidden patterns, structures, or relationships within the data on its own.

Think of it like handing a child a bucket of mixed blocks without instructions. They might naturally start grouping them by color, shape, or size. The child isn't "told" what a square is; they simply recognize that some blocks look more alike than others.

Common use cases include:

* **Clustering:** Grouping similar data points (e.g., customer segmentation).
* **Association:** Finding rules that describe your data (e.g., "people who buy onions also buy potatoes").
* **Dimensionality Reduction:** Simplifying data without losing its important trends.

# K-Means Clustering

**K-Means** is one of the most popular and straightforward clustering algorithms. Its goal is to partition $n$ observations into $K$ clusters, where each observation belongs to the cluster with the nearest mean (center point).

### The Step-by-Step Process

To perform K-Means, you follow these iterative steps:

1. **Choose the value of $K$:**
Decide how many clusters you want to find. For example, if $K=3$, the algorithm will try to find three distinct groups.
2. **Initialize Centroids:**
Randomly select $K$ data points from the dataset to serve as the initial **centroids** (the center of the clusters).
3. **Assignment Step:**
For every data point in your dataset, calculate its distance (usually Euclidean distance) to each of the $K$ centroids. Assign the data point to the cluster of the **nearest centroid**.
4. **Update Step (Re-centring):**
Now that the clusters are formed, calculate the new mean (average) of all the points in each cluster. Move the centroid of each cluster to this new mean position.
5. **Repeat:**
Repeat steps 3 and 4 until:
* The centroids no longer change their positions.
* The assignments of points to clusters remain the same.
* A maximum number of iterations is reached.



### How to Choose $K$?

Since we don't always know how many groups exist, we often use the **Elbow Method**. By plotting the "Within-Cluster Sum of Squares" (WCSS) against different values of $K$, you look for the "elbow" point where adding another cluster doesn't significantly improve the fit.

# Elbow Method

The **Elbow Method** is a heuristic used in cluster analysis to find the optimal number of clusters ($K$) in a dataset. It helps you strike a balance between having enough clusters to group data accurately without "overfitting" by having too many.

### The Core Concept: WCSS

To understand the Elbow Method, you first need to understand **WCSS** (Within-Cluster Sum of Squares).

* **WCSS** measures the compactness of the clusters.
* It is calculated by taking the sum of the squared distances between each data point and its assigned cluster centroid.
* **The Goal:** As $K$ (the number of clusters) increases, the WCSS value will naturally decrease because the points will be closer to their respective centroids.

---

### How the Method Works

1. **Run K-Means multiple times:** Execute the K-Means algorithm for a range of $K$ values (e.g., $K = 1$ to $10$).
2. **Calculate WCSS:** For each value of $K$, calculate the total WCSS.
3. **Plot the Graph:** Create a line screen where the x-axis represents the **Number of Clusters ($K$)** and the y-axis represents the **WCSS**.
4. **Identify the "Elbow":** As you increase $K$, the WCSS drops significantly at first. However, at a certain point, the rate of decrease slows down drastically, creating a sharp bend in the graph.

**The "Elbow Point" is the value of $K$ where this bend occurs.** This point represents the optimal tradeoff between error reduction and model complexity.

---

### Why not just choose the lowest WCSS?

If you have $n$ data points and you set $K = n$, every point becomes its own cluster. The WCSS would be **zero**, but the model would be useless because it hasn't actually "grouped" anything. The Elbow Method helps you find the point of diminishing returns.

### Limitations

While effective, the Elbow Method isn't always perfect:

* **Smooth Curves:** Sometimes the graph is a smooth curve without a clear "bend," making it subjective.
* **Alternative Methods:** In cases where the elbow is unclear, data scientists often use the **Silhouette Suite** or **Gap Statistic** to confirm the best $K$.


# Disadvantages

While K-Means is a "classic" algorithm for a reason—it’s fast and efficient—it has some significant "blind spots" that can lead to poor results if your data isn't perfectly shaped.

---

### 1. You Must Choose "K" in Advance

K-Means doesn't know how many clusters are in your data. You have to tell it (e.g., $k=3$). If you pick the wrong number, the algorithm will still force the data into those groups, even if it doesn't make sense.

* **The Fix:** Usually, you have to run an "Elbow Method" or "Silhouette Analysis" to find the best $k$.

### 2. Sensitive to Outliers

K-Means uses the **mean** (average) to find the center of a cluster. Just like one billionaire in a room of middle-class people makes the "average" wealth look huge, a single outlier can pull the cluster center (centroid) far away from the actual data.

### 3. Struggles with Non-Spherical Shapes

K-Means assumes that clusters are **spherical** (round) and roughly the same size.

* If your data looks like two moons or elongated ovals, K-Means will try to "cut" them into circles, leading to incorrect grouping.
* It also fails when clusters have very different densities (one cluster is packed tight, the other is spread out).

### 4. Random Initialization Trap

The algorithm starts by placing centroids in random spots. Sometimes, a "bad" starting position leads to a "sub-optimal" solution.

* **The Fix:** Scikit-learn uses `init='k-means++'` by default to pick smarter starting points, but it's still a factor to watch out for.

### 5. The "Curse of Dimensionality"

As you add more features (columns), the "distance" between points becomes less meaningful. In very high-dimensional space, everything starts looking like it's the same distance apart, which confuses the algorithm.

---

### When to use something else?

| If your data is... | Use this instead |
| --- | --- |
| Full of outliers | **K-Medoids** (uses the median instead of mean) |
| Oddly shaped (moons/circles) | **DBSCAN** or **Spectral Clustering** |
| Hierarchical in nature | **Agglomerative Clustering** |
